#### 3.字符串切片与下标

字符串编码，本质上是二进制与语言文字的一一对应关系

|编码|指定时间|作用|所占字节数|
|:--:|:--:|:--:|:--:|
|ASCII|1976年|表示英语和西欧语言|8bit/1bytes|
|GB2312|1980年|国家简体中文字符集，兼容ASCII|2bytes|
|Unicode|1991年|国际标准组织统一标准字符集|2bytes|
|GBK|1995年|GB2312的扩展字符集，支持繁体字，兼容GB2312|2bytes|
|**UTF-8**|1992年|不定长编码|1-3bytes|
---

- Unicode：所有字符都是两个字节，好处是字符与数字之间转换很快，坏处是占用空间大

- UTF-8：更加精准，对不同的长度字符用不同的长度表示，好处是节省空间，坏处是字符与数字转换速度较慢，每次都要计算字符用多少个字节表示

**字符串编码转换**

编码 `encode()` 将其他编码的字符串转换成Unicode编码

解码 `decode()` 将Unicode编码转换成其他编码的字符串

````python
a = "Hello"
print(a,type(a)) #type()函数可以查看变量类型

b = a.encode()
print("编码后：",b,type(b)) #encode()函数可以将字符串转换为字节类型

c = b.decode()
print("解码后：",c,type(c)) #decode()函数可以将字节类型转换为字符串类型
````

````python
str = "我江湖人称河地大刘德华"
st1 = str.encode("utf-8") #将字符串转换为字节类型
print(st1,type(st1)) #查看转换后的类型

st2 = st1.decode("utf-8") #将字节类型转换为字符串
print(st2,type(st2)) #查看转换后的类型
````

#### 5.字符串常见操作

1. `find` 检测某个字符串是否包含于另一字符串中，返回布尔

2. `count` 返回出现的次数

3. `replace` 替换

4. `split` 分割，以列表的形式返回，如果字符串中不包含分割内容，则不进行分割，返回整体字符串

5. `index` 和find比较像但是会报异常

6. `capitalize` 第一个字符大写

7. `startswith` 是否以某字符开头，返回布尔

`find` 检测某个字符串是否包含于另一字符串中，如果在就返回这个子字符串的下标，否则就返回-1

`find(子字符串，开始位置下标，结束未知下标)` 开始位置下标和结束未知下标可以省略，表示在整个字符串中查找

````python
name = "oliver Jackson"

print(name.find('r')) #第一个r的下标为5
print(name.find('Ja')) #区分大小写，是第一个字母所在的位置，J的下标为7
print(name.find('o',3)) #找的是从第3个开始往后找的
print(name.find('r',3,8)) #在第3到第8个中间找，不包含第八个，左闭右开区间
````

`index` 和 `find` 较像但是会报异常，如果在就返回子字符串开始位置下标，否则报错。

````python
name = 'oliver jackson'

print(name.index('i'))
print(name.index('o',2,13)) #未找到就报错，也是左开右闭区间
````

`count` 返回出现的次数，某个子字符串在整个字符串中出现的次数，没有就返回0

````python
name = 'oliver jackson'

print(name.count('o'))
print(name.count('o',2))
````

````python
print('oliver'.startswith('s')) #是否以某个字符开头

print('oliver'.endswith('r')) #是否以某个字符结尾

print('oliver'.upper()) #全部大写

print('JACKSOn'.lower()) #全部小写

print('JACKSOn'.capitalize()) #首字母大写，其他小写

print('oliver JACKSOn'.title()) #每个单词首字母大写，其他小写

print('oliver JACKSOn'.swapcase()) #大小写互换

print('Oliver'.isupper()) #是否全部大写
````

列表的常见操作

````python
b = 250
a = [6,"您好",True,b,48]

a.append("Jackson") #末尾加，整体添加
print(a)

a.extend('oli') #分散添加，将另一个类型的元素逐一添加，不可用于int，要可迭代对象
print(a)

a.insert(1,"早上好") #加载第二个元素前面，需要指定添加位置，原有元素后移
print(a)

a.pop(3) #去除第四个元素，Python3无参数时候会默认删除最后一个，指定的是索引，索引必须合理
print(a)

a.remove(48) #去除48这个元素，如果有重复的，默认删除第一个；列表必须包含元素，否则报错
print(a)

a.clear() #清空列表
print(a)

del a #删除列表
````

#### 2.深浅拷贝

````python
li = [1,2,3,4,5]
print(li,type(li)) #列表类型
li2 = li #引用了li的内存地址，li2和li指向同一块内存区域
print(li2,type(li2)) #列表类型
````

**浅拷贝**：会创建新的对象，拷贝第一层的数据，嵌套层会指向原来的额内存地址

优点：拷贝速度快，占用空间小，拷贝效率高

````python
import copy #导入copy模块

li = [1,2,3,[4,5,6],[7,8,9]]

li2 = copy.copy(li) #浅拷贝，li2和li指向不同的内存区域，但是li2的元素还是指向li的内存地址
print(li2,type(li2)) #列表类型

print("查看内存地址：",id(li))
print("查看内存地址：",id(li2)) #内存地址不一样，说明不是同一个对象

li.append(10) #添加元素，li2不会受到影响
print(li) 
print(li2) #没有受到影响

li[3].append(11) #嵌套列表添加元素，li2会受到影响，因为li2的元素还是指向li的内存地址
print(li,id(li[3])) #查看内存地址
print(li2,id(li2[3])) #受到影响，因为两个列表嵌套列表的元素内存地址是一样的
````

**深拷贝**：数据完全不共享，外层对象和内部元素都拷贝了一遍，数据变化只影响自己本身，与原有对象没有关系了

````python
import copy #导入copy模块

li = [1,2,3,[4,5,6],[7,8,9]]
li2 = copy.deepcopy(li) #深拷贝，li2和li指向不同的内存区域，li2的元素也指向不同的内存地址
print(li2,type(li2)) #列表类型
print("查看内存地址：",id(li))
print("查看内存地址：",id(li2)) #内存地址不一样，说明不是同一个对象

li.append(10) #添加元素，li2不会受到影响
print(li,"内存地址：",id(li)) #查看内存地址
print(li2,"内存地址：",id(li2)) #没有受到影响

li[3].append(11) #嵌套列表添加元素，li2不会受到影响，因为li2的元素也指向不同的内存地址
print(li,"内存地址：",id(li[3])) #查看内存地址
print(li2,"内存地址：",id(li2[3])) #没有受到影响，因为两个列表嵌套列表的元素内存地址是不一样的
````

#### 5.作用域与内置函数

**定义**：指的是变量生效的范围，分别是**全局变量**和**局部变量**

**全局变量**：函数外部定义的变量，在整个文件中都是有效的，函数内使用变量会优先从函数内部找，没有再从函数外找

````python
a = 10
def test1():
    print(a) #访问全局变量a

def test2():
    a = 20 #局部变量a，和全局变量a不冲突
    print(a) #访问局部变量a

print(a) #访问全局变量a
test1() #调用test1函数，访问全局变量a
test2() #调用test2函数，访问局部变量a
print(a) #访问全局变量a，局部变量a不影响全局变量a的值
````

**局部变量**：函数内部定义的变量，从定义位置开始到函数定义位置结束有效，只能在被定义的函数内部使用，不能在函数外部使用，在函数体内临时保存变量，调用完成后销毁

````python
def f1():
    n = 10
    print(n) #访问局部变量n

f1() #调用f1函数，访问局部变量n

def f2():
    n = 20 #局部变量n，和全局变量n不冲突
    print(n) #访问局部变量n

f2() #调用f2函数，访问局部变量n，两个函数的局部变量命名不冲突
````

在函数内部修改全局变量的值，使用 `global` 关键字，也可以在局部作用域中声明一个全局变量

````python
day = 11

def DAY():
    global day  #函数使用外部变量并且对外部变量进行了修改，就需要在函数最开始申明global
    day = 13 #局部变量day，和全局变量day不冲突

    if day % 2 == 0:
        print(f"吃饭{day}次")
    else:
        print(f"睡觉{day}次")
        day += 1

DAY()
print(day)
````

含有全局变量声明的函数必须调用后才可改变全局变量

````python
def Study():
    global name #申明name是全局变量
    name = "Oli" #局部变量name，和全局变量name不冲突
    print(name) #访问局部变量name

Study() #调用Study函数，访问局部变量name
print(name) #访问全局变量name，局部变量name不影响全局变量name的值
````

`nonlocal` 将变量声明为外层变量（外层函数的局部变量，而且不能是全局变量），对上一层进行修改

````python
a = 10

def test1():
    a = 20 #局部变量a，和全局变量a不冲突

    def test2():
        nonlocal a
        a = 30
        print("2函数中的值",a)
    test2() #调用test2函数，访问局部变量a
    print("1函数中的值",a) #访问局部变量a，此时的a已经被test2函数修改了

test1() #调用test2函数，访问局部变量a
print(a) #访问全局变量a，局部变量a不影响全局变量a的值
````

**匿名函数**

语法：函数名=lambda 形参：返回值

`lambda` 是定义匿名函数关键字，相当于 `def`

调用：结果 = 函数名（实参）

````python
def add(a,b): #定义函数，a和b是参数
    return a + b #返回a和b的和

print(add(1,2)) #调用函数，传入两个参数，返回值为a和b的和
````

匿名函数可以简化函数，不需要return来接受返回值，因为表达式本身就是返回值

````python
add = lambda a,b: a + b #匿名函数，a和b是参数，返回a和b的和
print(add(1,2)) #调用函数，传入两个参数，返回值为a和b的和

fun = lambda name:name
print(fun("Oliver")) #调用函数，传入一个参数，返回值为name的值

fun1 = lambda name,age = 18:(name,age) #匿名函数，name和age是参数，返回name和age的元组
print(fun1("Oliver")) #调用函数，传入一个参数，返回值为name的值
print(fun1("Oliver",23)) #调用函数，传入两个参数，返回值为name和age的元组
````

#### 1.递归与闭包

**含义**：如果一个函数在内部不调用其他函数，而是调用它本身的话，这个函数就是递归函数

**条件**：

1. 明确的结束条件

2. 每进行更深一层的递归时，问题规模相比上次递归都要有所减少

3. 相邻两次重复之间有紧密的联系

````python
def add():
    s = 0
    for i in range(1,101):
        s += i
    print(s)

add()
````

- **优点**：简洁逻辑清晰，解题更加具有思路

- **缺点**：使用递归函数的时候，需要反复调调用函数，耗内存，运行效率低

可以使用循环解决就不要用递归

````python
def add1(n):
    if n == 1:
        return 1 #如果不是1，重复执行累加并返回结果
    return n +add1(n-1)

print(add1(100))
````

斐波拉契数列

````python
def Fblq(n):
    if n <= 1:
        return n
    return Fblq(n-2) + Fblq(n-1)

print(Fblq(7))
````

**闭包**

含义：在嵌套函数的前提下，内部函数使用外部函数变量，而且外部函数返回了内部函数，就把使用外部函数变量的内部函数称为闭包

条件：

1. 嵌套函数（在函数里面再定义函数）

2. 内层函数使用外层函数的局部变量

3. 外层函数的返回值是内层函数的函数名

````python
def outer(): #外层函数
    n = 10
    def inner(): #内层函数
        print(n)
    return inner

print(outer()) #返回的是内部函数的内存地址

ot = outer() #调用外函数
ot() #调用内函数
````

调用外函数时，outer返回了inner，inner是函数的引用，存入了ot中，接着再使用ot( )就调用内函数

````python
def outer(m): #形参，外函数的局部变量
    n = 10
    def inner(o):
        print("计算结果：", m + n + o)
    return inner #返回函数名，而不是inner()，因为inner函数里面参数比较多的时或者受到限制的时，写法不太规范

out = outer(20)

out(10)
````

每次开启内函数都在使用同一份闭包变量

````python
def outer(m):
    print('outer()函数中的值',m)

    def inner(n):
        print('inner()函数中的值',n)
        return m + n
    
    return inner

ot = outer(10) #外函数的额10
print(ot(20)) #内函数的20
````

#### 2.装饰器

**含义**：装饰器本质上是一个Python函数，它可以让其他函数在不需要做任何代码变动的情况下增加函数功能，装饰器的返回值也是一个函数对象

装饰器需要满足以下两点

1. 不修改原程序或函数的代码

2. 不改变函数或程序调用的方法

````python
def test2():
    print('补考') #添加了一个补考

def test(fn):
    print('期末考试')
    print('没过')
    fn()  #调用要传入的函数

test(test2)
````

装饰器本质上就是一个闭包函数，原理是：将原有函数名定义为以原函数为参数的闭包

标准版装饰器：

````python
def exam():
    print('考100')

def outer(fn):
    
    def inner():
        print('参加期末') #被装饰的函数
        fn()
    return inner

ot = outer(exam)
ot()
````

**推荐用法**

Python提供更加简单的办法进行装饰，语法糖。

格式：@装饰器名称

````python
def outer(fn):
    def inner():
        print('登录教务系统...')
        fn()
    return inner

@outer  #被装饰的语法上面写个这个玩意儿，不要加入小括号，加了是调用函数，返回该函数返回的值
def send():
    print('登陆成功')

send()
````

被装饰函数有参数的情况

````python
def outer(fn):
    def inner(name):
        print(f"{name}位于名单中")
        fn(name)
    return inner

@outer
def func(name):
    print("-名单统计")
func('左恒滔')

# 内函数传递的参数和被装饰函数一致
````

被装饰的函数有可变参数，*args，**kwargs

````python
def func(*args,**kwargs): #被装饰函数
    print(args)
    print(kwargs)
# func(name = "Oliver")

def outer(fn):
    def inner(*args,**kwargs):
        print(fn)
        fn(*args,**kwargs)
    return inner

ot = outer(func)

ot('Oli',name = "Oliver")
````

**多个装饰器**

多个装饰器的装饰过程，离函数最近的装饰器先装饰，然后外面装饰器再进行装饰

````python
#第一个装饰器
def deco1(fn):
    def inner():
        return "哈哈" + fn() + "嘿嘿"
    return inner

# 第二个装饰器
def deco2(fn):
    def inner():
        return "很好" + fn() + "不错"
    return inner

#第三个装饰器
def deco3(fn):
    def inner():
        return "这是最外层" + fn() + "这是最外层"
    return inner


@deco3
@deco1
@deco2 #离得近的先装饰
#被装饰的函数1
def test1():
    return "期中考试"
print(test1())
#相当于"哈哈" + test1() + "嘿嘿"，fn()调用的时候相当于test1()在调用


````

#### 4.实例方法和实例属性

**实例方法**：由对象调用，至少有一个self参数，执行实例方法的时候，自动调用该方法的对象赋值给self

````python
class Washer:
    height = 800
    def wash(self): #self参数是类中的实例方法必须具备的
        print("洗衣服")
        print('方法中的：',self) #self表示当前调用方法的对象

#实例化对象
wa = Washer()
#对象调用类中的方法
wa.wash()

print(wa) #与self一样的
````

**实例属性**

格式：self.属性名

````python
class Person:
    name = "Oliver Jackson" #类名
    def run(self):      #self代表对象本身
        print("run方法中的self",self)
        print("人类会跑步")



#实例化对象：对象名 = 类名（）
pe = Person()
print(pe)
pe.run()
````

**实例属性和类属性区别**：

- **类属性**属于类，是公共的，大家都能访问到

- **实例属性**属于对象，是私有的，只能由对象名访问，不能由类名访问

````python
class Person:
    name = 'Oliver Jackson'
    def introduce(self): #实例方法
        print("这是实例方法")
        print(f"{Person.name}'age:{self.age}") #self.age是实例属性
              
pe = Person()
pe.age = 23
pe.weight = 65 
pe.introduce()

#访问类属性
print(Person.name)
print(pe.name)

pe2 = Person()
print(pe2.name)
#pe.age是给pe对象新增的实例属性，其他对象依然是没有这个属性
````

#### 8.静态方法与类方法

**静态方法**@staticmethod

既不需要传递类对象，也不需要传递实例对象（*没有self,cls参数的限制*）

静态方法与类无关，可以被转换成函数使用

```
class 类名：
    @staticmethod
    def 方法名（形参）：
        方法名
```
调用格式：
```
类名.方法名（实参）
对象名.方法名（实参）
```

````python
class Person(object):
    @staticmethod
    def study():
        print('人类会学习') #静态方法既可以使用对象访问，也可以使用类访问

Person.study()
pe = Person()
pe.study()
````

````python
class Person(object):
    @staticmethod
    def study(name):
        print(f'{name}会学习')

Person.study('Oliver') #调用方法时传参
pe = Person()
pe.study('Oli')
````

取消不必要的参数传递，有利于减少不必要的内存占用和性能消耗；不需要访问实例对象或者实例属性，或创建实例的时候就可以用。

使用装饰器@classmethod来标识为类方法，对于类方法，第一个参数必须是类对象，一般是以cls作为第一个参数
```
class 类名：
    @classmethod
    def 方法名（cls，形参）：
        方法体
```

### 四、示例

以下代码作为RSEI模型计算的演示，与实际计算RSEI的指数有一些差别，没有大气校正、去云等预处理操作，只是一个演示，不可以直接用在您的科研里

**注意：**

- 提前搞清楚使用什么包可以方便您的实验，不要想着什么都去自己写，您也写不出来PCA，当然我这里作为彩蛋提供原理[PCA原理](https://zhuanlan.zhihu.com/p/37777074)，您可以试着手撕代码，也就是按着数学原理来重现PCA算法

- 针对RSEI模型，一定搞清楚不同传感器TCT的coefficients，一点要参考官方文档，例如Landsat8系数参考[耶鲁大学-Tasseled Cap Transform for Landsat (8) OLI](https://yceo.yale.edu/tasseled-cap-transform-landsat-8-oli)，我在这上面吃了大亏，曾经图方便在一些第三方知识论坛上找，结果找到了错误的coefficients，给我的老师添了不少堵

- 您可以看看干度计算[计算干度（NDBSI）](https://blog.csdn.net/guaicaicexu/article/details/125539952)

- 记得打断点分布调试，去查看变量的变化情况

````python
import numpy as np
import time
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

time_start = time.time()

#这里使用random.uniform生成随机的float32数，注意random.randint就不可以
data = np.random.uniform(low=0.0, high=1.0, size=(2000, 2000, 7)).astype(np.float32)

# 我假装是Landsat 8，来参考缨帽系数
tct_coefficients = {
    'brightness': [0.3029, 0.2786, 0.4733, 0.5599, 0.5080, 0.1872],
    # 'greenness': [-0.2941, -0.2430, -0.5424, 0.7276, 0.0713, -0.1608], #我们这里应该参考徐涵秋老师的公式，不用缨帽的绿度
    'wetness': [0.1511, 0.1973, 0.3283, 0.3407, -0.7117, -0.4559]
}

# 提取前6个波段（B2-B7）进行缨帽变换
bands_tct = data[..., :6]

# 计算缨帽分量（绿度、亮度、湿度）,这个地方的绿度分量我们不用缨帽变换，单独用NDVI提出来算
brightness = np.sum(bands_tct * tct_coefficients['brightness'], axis=2)
greenness = (data[...,4] - data[...,3])/(data[...,4] + data[...,3] + 1e-10)
wetness = np.sum(bands_tct * tct_coefficients['wetness'], axis=2)

# 干度指数 = (SWIR1 - 近红外) / (SWIR1 + 近红外)
#我想偷懒没偷成，不要学我，一开始就得标注好波段索引
CA = 0      #B1
BLUE = 1    # B2
GREEN = 2   # B3
RED = 3     # B4
NIR = 4     # B5
SWIR1 = 5   # B6
SWIR2 = 6   # B7

def safe_divide(a, b):
    """安全的除法运算，防止除以零"""
    return a / (b + 1e-10)

# 计算IBI分量
SWIR1_band = data[..., SWIR1]
NIR_band = data[..., NIR]
RED_band = data[..., RED]
GREEN_band = data[..., GREEN]
BLUE_band = data[..., BLUE]

IBI_numerator = 2 * safe_divide(SWIR1_band, SWIR1_band + NIR_band) - \
               (safe_divide(NIR_band, NIR_band + RED_band) + 
                safe_divide(GREEN_band, GREEN_band + SWIR1_band))

IBI_denominator = 2 * safe_divide(SWIR1_band, SWIR1_band + NIR_band) + \
                 (safe_divide(NIR_band, NIR_band + RED_band) + 
                  safe_divide(GREEN_band, GREEN_band + SWIR1_band))

IBI = safe_divide(IBI_numerator, IBI_denominator)

# 计算SI分量
SI_numerator = (SWIR1_band + RED_band) - (NIR_band + BLUE_band)
SI_denominator = (SWIR1_band + RED_band) + (NIR_band + BLUE_band)
SI = safe_divide(SI_numerator, SI_denominator)

# 计算NDBSI
dryness = (IBI + SI) / 2

# 地表温度（模拟一下热红外B10，我们假装温度范围280-320K归一化到0-1），但是真实的地表温度不可以这么算，要参考正经的LST指数，我这里偷个懒
lst = data[...,6] * 40 + 280  # 反归一化到温度范围

# 标准化（这里具体应该标准化还是直接归一化我忘了，留给您自己查叭）
scaler = MinMaxScaler()
indicators = np.stack([
    scaler.fit_transform(greenness.reshape(-1,1)).flatten(),  # 绿度
    scaler.fit_transform(wetness.reshape(-1,1)).flatten(),    # 湿度
    scaler.fit_transform(dryness.reshape(-1,1)).flatten(),    # 干度
    scaler.fit_transform(lst.reshape(-1,1)).flatten()         # 热度
], axis=1)# 列方向

# PCA
pca = PCA(n_components=1)
rsei = pca.fit_transform(indicators).reshape(2000, 2000)

# 归一化
rsei_norm = (rsei - rsei.min()) / (rsei.max() - rsei.min())

# 可视化，因为是随机数，所以结果就是均匀的颜色，是正常的
plt.figure(figsize=(15,6))
# plt.subplot(141), plt.imshow(greenness, cmap='YlGn'), plt.title('TCT Greenness')
plt.subplot(142), plt.imshow(dryness, cmap='Blues'), plt.title('NDBSI')
plt.subplot(143), plt.imshow(rsei_norm, cmap='RdYlGn_r'), plt.title('RSEI')
plt.subplot(144),plt.imshow(greenness,cmap = 'YlGn'),plt.title('NDVI')
plt.colorbar(orientation='horizontal'), plt.tight_layout(), plt.show()


useage = (time.time() - time_start)
print(f'计算完成，总共耗时：{useage:.2f}秒')
````

我把上面这个一样的RSEI模型转移到GPU上进行计算，因为它有矩阵运算能力强的特点（如果您是用的苹果的Mac那么您可能需要通过Accelerate间接调用M系列处理器封装的AMX矩阵运算单元），且有不少针对的加速包。您可以看到效率好像没有提升，这只是针对2000*2000且只有一张影像的这种小规模（迷您规模）的计算，要是在使用PCA（针对性优化）而不是SVD，并且规模很大，长时间序列或者面积特别大的情况下，提升会更大。不过我现在觉得还是写上面的CPU版就行，简单高效，且学习成本很低。

**注意**

- SVD的数学表达式为：$$A = U \Sigma V^T$$ 其中，A是原始矩阵，U和V是正交矩阵，Sigma是对角矩阵，包含奇异值。

- PCA的数学表达式为：$$X = U \Sigma V^T$$ 其中，X是原始数据矩阵，U是左奇异矩阵，Sigma是奇异值矩阵，V是右奇异矩阵。

这个代码看看就行，其实有很多地方做的不妥，因为一旦涉及到数据到GPU上算，这个程序设计需要考虑的方面就稍微多一些，我这方面做的还不是很好，不过希望能给您对于程序优化的参考

````python
import torch
import time
import numpy as np
import matplotlib.pyplot as plt

torch.cuda.empty_cache()

# 检查GPU是否可用（这里强调必须是英伟达的显卡而非AMD，AMD得使用OpenCL的一些社区支持才能用CUDA，这个我没有了解过）
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

time_start = time.time()
lo_we = 2000

# 随机数据（直接在GPU创建）
data = torch.rand((lo_we, lo_we, 7), device=device, dtype=torch.float32)  # 1000x1000x7

# 缨帽系数，我这里偷个懒，不想写NDVI了，直接缨帽一个绿度（GPU张量）
tct_coefficients = {
    'brightness': torch.tensor([0.3029, 0.2786, 0.4733, 0.5599, 0.5080, 0.1872], device=device),
    'greenness': torch.tensor([-0.2941, -0.2430, -0.5424, 0.7276, 0.0713, -0.1608], device=device),
    'wetness': torch.tensor([0.1511, 0.1973, 0.3283, 0.3407, -0.7117, -0.4559], device=device)
}

# 提取前6波段（B2-B7）
bands_tct = data[..., :6]

#  缨帽变换（GPU）
brightness = (bands_tct * tct_coefficients['brightness']).sum(dim=2)
greenness = (bands_tct * tct_coefficients['greenness']).sum(dim=2)
wetness = (bands_tct * tct_coefficients['wetness']).sum(dim=2)

CA = 0      #B1
BLUE = 1    # B2
GREEN = 2   # B3
RED = 3     # B4
NIR = 4     # B5
SWIR1 = 5   # B6
SWIR2 = 6   # B7

def safe_divide(a, b):
    """安全的除法运算，防止除以零"""
    return a / (b + 1e-10)

# 计算IBI分量
SWIR1_band = data[..., SWIR1]
NIR_band = data[..., NIR]
RED_band = data[..., RED]
GREEN_band = data[..., GREEN]
BLUE_band = data[..., BLUE]

IBI_numerator = 2 * safe_divide(SWIR1_band, SWIR1_band + NIR_band) - \
               (safe_divide(NIR_band, NIR_band + RED_band) + 
                safe_divide(GREEN_band, GREEN_band + SWIR1_band))

IBI_denominator = 2 * safe_divide(SWIR1_band, SWIR1_band + NIR_band) + \
                 (safe_divide(NIR_band, NIR_band + RED_band) + 
                  safe_divide(GREEN_band, GREEN_band + SWIR1_band))

IBI = safe_divide(IBI_numerator, IBI_denominator)

# 计算SI分量
SI_numerator = (SWIR1_band + RED_band) - (NIR_band + BLUE_band)
SI_denominator = (SWIR1_band + RED_band) + (NIR_band + BLUE_band)
SI = safe_divide(SI_numerator, SI_denominator)

# 计算NDBSI
dryness = (IBI + SI) / 2

# 温度数据（GPU）
lst = data[...,6] * 40 + 280  # 反归一化到280-320K

#  数据标准化，这个没有现成的方法，反正我不知道有没有，那就自己手算一下标准化，很简单的
def gpu_minmax_scale(tensor):
    t_min = tensor.min()
    t_max = tensor.max()
    return (tensor - t_min) / (t_max - t_min)

greenness_norm = gpu_minmax_scale(greenness).view(-1, 1)
wetness_norm = gpu_minmax_scale(wetness).view(-1, 1)
dryness_norm = gpu_minmax_scale(dryness).view(-1, 1)
lst_norm = gpu_minmax_scale(lst).view(-1, 1)

# 合并指标并转置为[样本数, 特征数]
indicators = torch.cat([greenness_norm, wetness_norm, dryness_norm, lst_norm], dim=1)

# 手动PCA，这个是基于SVD分解，这部分我还没有学会，因为好像我还不知道有没有cuda支持的PCA方法，所以这部分其实是GPT提示我用SVD来替代PCA，我并没有研究用法，不过您要知道PCA和SVD数学原理其实是不太一样的，虽然都是完成数据主要特征提取
mean = torch.mean(indicators, dim=0)
X_centered = indicators - mean
U, S, V = torch.linalg.svd(X_centered, full_matrices=False)
rsei = (X_centered @ V[:, :1]).view(lo_we, lo_we)

# 归一化
rsei_norm = (rsei - rsei.min()) / (rsei.max() - rsei.min())

# 可视化（这个要传回CPU）
plt.figure(figsize=(15,6))
plt.subplot(131), plt.imshow(greenness.cpu(), cmap='YlGn'), plt.title('TCT Greenness(not NDVI ,I\'m being lazy)')
plt.subplot(132), plt.imshow(dryness.cpu(), cmap='Blues'), plt.title('NDBSI')
plt.subplot(133), plt.imshow(rsei_norm.cpu(), cmap='RdYlGn_r'), plt.title('RSEI')
plt.colorbar(orientation='horizontal'), plt.tight_layout(), plt.show()

# 计时
useage = time.time() - time_start
print(f'计算完成，总共耗时：{useage:.2f}秒 | 设备：{device}')

#因为我没写成一共完整的函数，所以得手动清空一下显存
torch.cuda.empty_cache()
````

````python
%%javascript

// 获取NDVI计算结果
async function getNDVI() {
  await new Promise(resolve => setTimeout(resolve, 500)); // 模拟服务器延迟
  return 0.72; 
}

(async () => {
  const ndvi = await getNDVI();
  element.textContent += `\n异步NDVI值: ${ndvi}`; 
})();
````

#### 4.错误处理与调试

**调试技巧**

多分步运算，且在个段单独设置控制台打印输出内容，这样方便检查数据类型哪里有不对的，比如归一化是否有误，栅格有误非常大或者非常小的明显的错误值

````python
%%javascript

// 波段值验证调试
function validateBand(band) {
  console.log('当前波段:', band);  // 浏览器控制台可见
  return band.startsWith('B') && band.length === 2;
}

const testBand = 'B10';
element.textContent = `波段${testBand}有效? ${validateBand(testBand)}`;
// 输出: 波段B10有效? true
````

**断言模式**

是编程中用于验证条件是否满足的防御性编程技术，当条件不成立时立即抛出错误。以下是专门针对遥感数据处理场景的详细解释

可以用在：

- 植被指数计算前验证输入波段
- 时间序列分析时检查日期范围

````python
%%javascript

// NDVI有效性检查
function assertNDVI(value) {
  if (value < -1 || value > 1) 
    throw new Error(`异常NDVI值: ${value}`);
  return value.toFixed(2);
}

try {
  assertNDVI(1.2);
} catch (e) {
  element.textContent = e.message;  // 输出"异常NDVI值: 1.2"
}
````
